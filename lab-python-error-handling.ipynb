### STEP 1
modify the `initialize_inventory` function to include error handling.
   - **If the user enters an invalid quantity** (e.g., a negative value or a non-numeric value), **display an error message** and **ask them to re-enter** the quantity for that product.
   - Use a **``try-except``** block to handle the error and continue prompting the user **until a valid quantity is entered**.

````python
# Step 1: Define the function for initializing the inventory with error handling
def initialize_inventory(products):
    """
    Initialize inventory quantities for a list of products.
    Ensures that the user enters valid non-negative integers.
    
    Args:
        products (list): List of product names.
    
    Returns:
        dict: Dictionary with product names as keys and quantities as values.
    """
    inventory = {}
    
    for product in products:
        while True:
            try:
                # Ask the user for quantity
                quantity = input(f"Enter quantity for {product}: ")
                
                # Convert input to integer
                quantity = int(quantity)
                
                # Check for negative values
                if quantity < 0:
                    raise ValueError("Quantity cannot be negative.")
                
                # Valid quantity, store it and break the loop
                inventory[product] = quantity
                break
            
            except ValueError as e:
                # Handle non-numeric or negative inputs
                print(f"Invalid input: {e}. Please enter a non-negative integer.")
    
    return inventory

# Test the function:
products_list = ["Apples", "Bananas", "Oranges"]
inventory = initialize_inventory(products_list)
print("Inventory initialized:", inventory)
````

### STEP 2
2. Modify the `calculate_total_price` function to include error handling.
   
   - If the user enters an invalid price (e.g., a negative value or a non-numeric value),
   - display an error message and ask them to re-enter the price for that product.
   
   - Use a try-except block to handle the error and continue prompting the user until a valid price is entered.

````python
def calculate_total_price(products):
    """
    Ask the user to enter the price for each product and calculate the total price.
    Ensures valid, non-negative numeric input.

    Args:
        products (list): List of product names.

    Returns:
        float: Total price of all products.
    """
    total_price = 0.0
    
    for product in products:
        while True:
            try:
                # Ask user for the price
                price = input(f"Enter price for {product}: ")
                
                # Convert input to float
                price = float(price)
                
                # Check for negative price
                if price < 0:
                    raise ValueError("Price cannot be negative.")
                
                # Valid price, add to total and break loop
                total_price += price
                break
            
            except ValueError as e:
                print(f"Invalid input: {e}. Please enter a non-negative number.")
    
    return total_price

# Example usage:
products_list = ["Apples", "Bananas", "Oranges"]
total = calculate_total_price(products_list)
print(f"Total price: ${total:.2f}")
````

### STEP 3
3. Modify the `get_customer_orders` function to include error handling.
   - If the user enters an invalid number of orders (e.g., a negative value or a non-numeric value), 
      - display an error message and ask them to re-enter the number of orders.
   - If the user enters an invalid product name (e.g., a product name that is not in the inventory), or that doesn't have stock available, 
      - display an error message and ask them to re-enter the product name. 
   - *Hint*: you will need to pass inventory as a parameter*
   - Use a `try-except` block to handle the error and continue prompting the user until a valid product name is entered.


````python
def get_customer_orders(inventory):
    """
    Ask the user for their orders, including quantity and product name.
    Validates:
      - Number of orders (non-negative integer)
      - Product exists in inventory
      - Product has available stock

    Args:
        inventory (dict): Dictionary of product names and available quantities

    Returns:
        dict: Customer orders with product names and ordered quantities
    """
    orders = {}

    print("\nEnter your orders:")

    while True:
        try:
            num_orders = input("How many different products do you want to order? ")
            num_orders = int(num_orders)
            if num_orders < 0:
                raise ValueError("Number of orders cannot be negative.")
            break
        except ValueError as e:
            print(f"Invalid input: {e}. Please enter a non-negative integer.")

    for i in range(num_orders):
        while True:
            product_name = input(f"Enter product name for order #{i+1}: ").strip()
            
            # Check if product exists in inventory
            if product_name not in inventory:
                print("Invalid product name. Please choose a product from the inventory.")
                continue
            
            # Check if product has stock available
            if inventory[product_name] == 0:
                print(f"Sorry, {product_name} is out of stock. Please choose another product.")
                continue
            
            # Ask for quantity
            while True:
                try:
                    quantity = input(f"Enter quantity for {product_name}: ")
                    quantity = int(quantity)
                    if quantity < 0:
                        raise ValueError("Quantity cannot be negative.")
                    if quantity > inventory[product_name]:
                        raise ValueError(f"Only {inventory[product_name]} units of {product_name} available.")
                    break
                except ValueError as e:
                    print(f"Invalid input: {e}. Please try again.")
            
            # Save order and reduce inventory
            orders[product_name] = quantity
            inventory[product_name] -= quantity
            break

    return orders

# Example usage
inventory = {"Apples": 10, "Bananas": 5, "Oranges": 8}
customer_orders = get_customer_orders(inventory)
print("\nCustomer Orders:", customer_orders)
print("Updated Inventory:", inventory)
````
