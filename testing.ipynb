```python
def initialize_inventory(products):
    inventory = {}
    for product in products:
        while True:
            try:
                quantity = int(input(f"Enter quantity of {product}s: "))
                if quantity >= 0:
                    inventory[product] = quantity
                    break
                print("Quantity cannot be negative!")
            except ValueError:
                print("Enter a valid number!")
    return inventory

def calculate_total_price(customer_orders):
    total_price = 0
    for product in customer_orders:
        while True:
            try:
                price = float(input(f"Enter price for {product}: $"))
                if price >= 0:
                    total_price += price * customer_orders[product]
                    break
                print("Price cannot be negative!")
            except ValueError:
                print("Enter a valid number!")
    return total_price

def get_customer_orders(inventory):
    customer_orders = {}
    
    while True:
        try:
            num_orders = int(input("How many different products? "))
            if num_orders > 0:
                break
            print("Number must be positive!")
        except ValueError:
            print("Enter a valid number!")
    
    for _ in range(num_orders):
        while True:
            product = input("What product? ").strip()
            if product in inventory and inventory[product] > 0:
                customer_orders[product] = 1
                break
            print(f"{product} not available!")
    
    return customer_orders

# Test it
products = ["t-shirt", "mug", "hat"]
inventory = initialize_inventory(products)
orders = get_customer_orders(inventory)
total = calculate_total_price(orders)
print(f"Total: ${total}")
```
